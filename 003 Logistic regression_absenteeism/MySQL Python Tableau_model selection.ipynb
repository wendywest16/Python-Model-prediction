```python
#Standardise the data
from sklearn.preprocessing import StandardScaler

#absenteeism_scaler = StandardScaler() #Used to subtract the mean and divide by the standard deviation variablewise

#Whenever you get new data you will know that the standardisation info is contained in the absenteeism_scaler. So you can
#standardise the new data in the same way

#But this scaler scales ALL columns, when we don't want the dummy variables (Reason 1,2,3,4, Education) to be scaled
```

```python
#Create a custom scaler so that the dummy variables (Reason 1,2,3,4) aren't standardised too. 
#It allows you to choose the variables that you'd like to scale.
from sklearn.base import BaseEstimator, TransformerMixin
class CustomScaler(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns, copy = True, with_mean=True, with_std=True):
        self.scaler = StandardScaler(copy,with_mean,with_std)
        self.columns = columns
        self.mean_ = None
        self_var_ = None
        
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns = self.columns)
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis = 1)[init_col_order]
```

```python
unscaled_inputs.columns.values
```

```python
columns_to_omit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']
```

```python
columns_to_scale = [x for x in unscaled_inputs.columns.values if x not in columns_to_omit] #excludes columns in the 
#columns_to_omit list
```

```python
absenteeism_scaler = CustomScaler(columns_to_scale) #This is the scaling mechanism
```

```python
absenteeism_scaler.fit(unscaled_inputs)
```

```python
scaled_inputs = absenteeism_scaler.transform(unscaled_inputs) #Transforms the unscaled inputs using information contained in
#absenteeism_scaler (i.e. subtract the mean and divide by the standard deviation variablewise)
```
